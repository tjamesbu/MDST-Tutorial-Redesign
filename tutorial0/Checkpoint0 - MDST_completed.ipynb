```python
# load in the "starbucks.csv" dataset
# refer to how we read the cereal.csv dataset in the tutorial
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/tjamesbu/MDST-Tutorial-Redesign/main/data/starbucks.csv')
data.head()
```

```python
data.columns
```

```python
# output the calories, sugars, and protein columns only of every 40th row.
cal_every_40_rows = data['calories'].loc[::40]
sugars_every_40_rows = data['sugars'].loc[::40]
protein_every_40_rows = data['protein'].loc[::40]
```

```python
# select all rows with more than and including 400 calories
more_than_400_calories = data[data.calories >= 400] 
more_than_400_calories
```

```python
# select all rows whose vitamin c content is higher than the iron content
vit_c_greater_than_iron = data[data['vitamin c'] > data['iron']]
vit_c_greater_than_iron
```

```python
# create a new column containing the caffeine per calories of each drink
data['caffeine_per_calories'] = data['caffeine']/data['calories']
data.head()
```

```python
# what is the average calorie across all items?
avg_cals = data['calories'].mean()
avg_cals
```

```python
# how many different categories of beverages are there?
bev_cat_count = len(data.beverage_category.unique())
bev_cat_count
```

```python
data.beverage_category.unique()
```

```python
# what is the average # calories for each beverage category?
coffee_data = data[data.beverage_category == 'coffee']
coffee_data_sum = coffee_data['calories'].sum()
coffee_data_len = len(coffee_data['calories'])
coffee_data_avg = round(coffee_data_sum/coffee_data_len)
#coffee_data_avg

ced_data = data[data.beverage_category == 'classic espresso drinks']
ced_data_sum = ced_data['calories'].sum()
ced_data_len = len(ced_data['calories'])
ced_data_avg = round(ced_data_sum/ced_data_len)

sed_data = data[data.beverage_category == 'signature espresso drinks']
sed_data_sum = sed_data['calories'].sum()
sed_data_len = len(sed_data['calories'])
sed_data_avg = round(sed_data_sum/sed_data_len)

ttd_data = data[data.beverage_category == 'tazo tea drinks']
ttd_data_sum = ttd_data['calories'].sum()
ttd_data_len = len(ttd_data['calories'])
ttd_data_avg = round(ttd_data_sum/ttd_data_len)

sib_data = data[data.beverage_category == 'shaken iced beverages']
sib_data_sum = sib_data['calories'].sum()
sib_data_len = len(sib_data['calories'])
sib_data_avg = round(sib_data_sum/sib_data_len)

smooth_data = data[data.beverage_category == 'smoothies']
smooth_data_sum = smooth_data['calories'].sum()
smooth_data_len = len(smooth_data['calories'])
smooth_data_avg = round(smooth_data_sum/smooth_data_len)

fbc_data = data[data.beverage_category == 'frappuccino blended coffee']
fbc_data_sum = coffee_data['calories'].sum()
fbc_data_len = len(coffee_data['calories'])
fbc_data_avg = round(coffee_data_sum/coffee_data_len)

fbc1_data = data[data.beverage_category == 'frappuccino light blended coffee']
fbc1_data_sum = fbc1_data['calories'].sum()
fbc1_data_len = len(fbc1_data['calories'])
fbc1_data_avg = round(fbc1_data_sum/fbc1_data_len)

fbc2_data = data[data.beverage_category == 'frappuccino blended crme']
fbc2_data_sum = fbc2_data['calories'].sum()
fbc2_data_len = len(fbc2_data['calories'])
fbc2_data_avg = round(fbc2_data_sum/fbc2_data_len)

dict1 = {'beverage_category': ['coffee', 'classic espresso drinks', 'signature espresso drinks', 'tazo tea drinks', 'shaken iced beverages', 'smoothies', 'frappuccino blended coffee', 'frappuccino light blended coffee', 'frappuccino blended crme'], 
         'average_calories': [coffee_data_avg, ced_data_avg, sed_data_avg, ttd_data_avg, sib_data_avg, smooth_data_avg, fbc_data_avg, fbc1_data_avg, fbc2_data_avg]}
df = pd.DataFrame(data=dict1)
df.head()
```
